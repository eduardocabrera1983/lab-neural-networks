```python
# your code here

# Spiral Data Classification with TensorFlow - Complete Solution

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from math import hypot, cos, sin

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("TensorFlow version:", tf.__version__)
print("Pandas version:", pd.__version__)
print("=== SPIRAL DATA NEURAL NETWORK CLASSIFICATION ===\n")

# Note: In newer pandas versions (1.4.0+), DataFrame.append() is deprecated
# We use pd.concat() instead for combining DataFrames

# ============================================================================
# STEP 0: GENERATE SPIRAL DATA (from the notebook)
# ============================================================================

def spiral(radius, step, resolution=.1, angle=0.0, start=0.0):
    """Generate X/Y data points that will form a spiral."""
    dist = start
    coords = []
    while dist * hypot(cos(angle), sin(angle)) < radius:
        cord = []
        cord.append(dist * cos(angle))
        cord.append(dist * sin(angle))
        coords.append(cord)
        dist += step
        angle += resolution
    return coords

# Generate two sets of spiral data points with opposite angles
print("Generating spiral datasets...")
data_1 = np.array(spiral(1000, 5, angle=0))
data_2 = np.array(spiral(1000, 5, angle=180))

# Create combined dataset
df1 = pd.DataFrame(data=data_1, columns=["X", "Y"])
df1["CLASS"] = 0

df2 = pd.DataFrame(data=data_2, columns=["X", "Y"])
df2["CLASS"] = 1

# Use pd.concat() instead of deprecated .append() method
df = pd.concat([df1, df2], ignore_index=True)
print("✓ Spiral data generated!")
print("Class distribution:")
print(df['CLASS'].value_counts())

# Quick visualization of the spiral data
plt.figure(figsize=(8, 6))
a, b = data_1.T
plt.scatter(a, b, s=5, alpha=0.7, label='Class 0')

aa, bb = data_2.T
plt.scatter(aa, bb, s=5, alpha=0.7, label='Class 1')

plt.title('Generated Spiral Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ============================================================================
# STEP 1: FEATURE ENGINEERING (Apply Playground Lessons)
# ============================================================================

def create_spiral_features(df):
    """
    Create polynomial and trigonometric features like in TensorFlow Playground
    Based on your successful playground configuration
    """
    print("Creating advanced features (like TensorFlow Playground)...")
    
    # Original features
    X1 = df['X'].values
    X2 = df['Y'].values
    
    # Polynomial features (essential for spiral patterns)
    X1_squared = X1 ** 2      # X₁² - captures curvature in X direction
    X2_squared = X2 ** 2      # X₂² - captures curvature in Y direction  
    X1_X2 = X1 * X2          # X₁X₂ - captures spiral interaction
    
    # Trigonometric features (helped in your playground solution)
    sin_X1 = np.sin(X1)      # sin(X₁) - periodic patterns
    sin_X2 = np.sin(X2)      # sin(X₂) - rotational patterns
    
    # Create feature matrix (similar to playground feature selection)
    features = np.column_stack([
        X1,          # X₁
        X2,          # X₂  
        X1_squared,  # X₁²
        X2_squared,  # X₂²
        X1_X2,       # X₁X₂
        sin_X1,      # sin(X₁)
        sin_X2       # sin(X₂)
    ])
    
    feature_names = ['X1', 'X2', 'X1²', 'X2²', 'X1*X2', 'sin(X1)', 'sin(X2)']
    
    print(f"✓ Created {features.shape[1]} features: {feature_names}")
    print(f"✓ Feature matrix shape: {features.shape}")
    
    return features, feature_names

# Create features from the spiral dataframe
X_features, feature_names = create_spiral_features(df)
y = df['CLASS'].values

print(f"✓ Total samples: {len(y)}")
print(f"✓ Class distribution: {np.bincount(y)}")

# ============================================================================
# STEP 2: DATA PREPROCESSING
# ============================================================================

print("\n--- Data Preprocessing ---")

# Train-test split (80-20 like playground)
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y  # Maintain class balance
)

print(f"✓ Training samples: {X_train.shape[0]}")
print(f"✓ Test samples: {X_test.shape[0]}")

# Feature scaling (essential for neural networks)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"✓ Features normalized (mean ≈ 0, std ≈ 1)")
print(f"  Training data range: [{X_train_scaled.min():.2f}, {X_train_scaled.max():.2f}]")

# ============================================================================
# STEP 3: BUILD NEURAL NETWORK (Replicate Playground Success)
# ============================================================================

print("\n--- Building Neural Network ---")
print("Replicating your successful TensorFlow Playground configuration...")

# Based on successful playground setup:
# - 3 hidden layers with [4, 5, 5] neurons
# - Sigmoid activation (your choice from playground)
# - Learning rate: 0.03

model = tf.keras.Sequential([
    # Input layer (7 features)
    tf.keras.layers.Dense(4, activation='sigmoid', input_shape=(7,), name='hidden_1'),
    
    # Hidden layers (matching your playground architecture)
    tf.keras.layers.Dense(5, activation='sigmoid', name='hidden_2'),
    tf.keras.layers.Dense(5, activation='sigmoid', name='hidden_3'),
    
    # Output layer (2 classes with softmax)
    tf.keras.layers.Dense(2, activation='softmax', name='output')
])

# Compile with your successful playground hyperparameters
custom_optimizer = tf.keras.optimizers.Adam(learning_rate=0.03)  # Your playground LR

model.compile(
    optimizer=custom_optimizer,
    loss='sparse_categorical_crossentropy',  # For integer labels (0, 1)
    metrics=['accuracy']
)

# Display model architecture
print("✓ Neural Network Architecture:")
model.summary()

# Count parameters
total_params = model.count_params()
print(f"✓ Total parameters: {total_params}")

# ============================================================================
# STEP 4: TRAINING (Monitor Like Playground)
# ============================================================================

print("\n--- Training Neural Network ---")
print("Training with playground-inspired hyperparameters...")

# Train the model (similar epoch count to your playground success)
history = model.fit(
    X_train_scaled, y_train,
    epochs=1200,                    # Reasonable epoch count
    batch_size=16,                 # Small batch size for stability
    validation_split=0.1,          # 10% validation data
    verbose=1,                     # Show progress
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss', 
            patience=20, 
            restore_best_weights=True
        )
    ]
)

print("✓ Training completed!")

# ============================================================================
# STEP 5: EVALUATION
# ============================================================================

print("\n--- Model Evaluation ---")

# Evaluate on test data
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test, verbose=0)
train_loss, train_accuracy = model.evaluate(X_train_scaled, y_train, verbose=0)

print(f"📊 FINAL RESULTS:")
print(f"   Training Loss: {train_loss:.4f}")
print(f"   Training Accuracy: {train_accuracy:.4f}")
print(f"   Test Loss: {test_loss:.4f}")
print(f"   Test Accuracy: {test_accuracy:.4f}")

# Check if we achieved playground-level performance
if test_loss < 0.05:
    print("🎉 SUCCESS: Test loss < 0.05 (matching playground goal!)") 
elif test_loss < 0.1:
    print("✅ GOOD: Test loss < 0.1 (close to playground goal)")
else:
    print("⚠️  ROOM FOR IMPROVEMENT: Consider adjusting architecture or hyperparameters")

# Detailed classification metrics
y_pred = model.predict(X_test_scaled)
y_pred_classes = np.argmax(y_pred, axis=1)

print(f"\n📋 Detailed Classification Report:")
print(classification_report(y_test, y_pred_classes))

# ============================================================================
# STEP 6: VISUALIZATION
# ============================================================================

print("\n--- Creating Visualizations ---")

# Plot training history
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Loss curves
ax1.plot(history.history['loss'], label='Training Loss', color='blue')
ax1.plot(history.history['val_loss'], label='Validation Loss', color='red')
ax1.set_title('Model Loss During Training')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Accuracy curves  
ax2.plot(history.history['accuracy'], label='Training Accuracy', color='blue')
ax2.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
ax2.set_title('Model Accuracy During Training')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3)
ax3.set_title('Confusion Matrix')
ax3.set_xlabel('Predicted')
ax3.set_ylabel('Actual')

# Decision boundary visualization (create a grid and predict)
def plot_decision_boundary():
    # Create a grid of points
    h = 0.5  # Grid spacing
    x_min, x_max = df['X'].min() - 1, df['X'].max() + 1
    y_min, y_max = df['Y'].min() - 1, df['Y'].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                        np.arange(y_min, y_max, h))
    
    # Create grid dataframe and engineer features
    grid_df = pd.DataFrame({
        'X': xx.ravel(),
        'Y': yy.ravel()
    })
    
    grid_features, _ = create_spiral_features(grid_df)
    grid_features_scaled = scaler.transform(grid_features)
    
    # Predict on grid
    grid_predictions = model.predict(grid_features_scaled)
    grid_predictions = np.argmax(grid_predictions, axis=1)
    grid_predictions = grid_predictions.reshape(xx.shape)
    
    return xx, yy, grid_predictions

xx, yy, grid_pred = plot_decision_boundary()

# Plot decision boundary with original data
ax4.contourf(xx, yy, grid_pred, alpha=0.3, cmap='RdYlBu')
scatter1 = ax4.scatter(df[df['CLASS']==0]['X'], df[df['CLASS']==0]['Y'], 
                      c='blue', s=20, alpha=0.7, label='Class 0')
scatter2 = ax4.scatter(df[df['CLASS']==1]['X'], df[df['CLASS']==1]['Y'], 
                      c='red', s=20, alpha=0.7, label='Class 1')
ax4.set_title('Decision Boundary Visualization')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.legend()

plt.tight_layout()
plt.show()

# ============================================================================
# STEP 7: FEATURE IMPORTANCE ANALYSIS
# ============================================================================

print("\n--- Feature Importance Analysis ---")

# Analyze which features the model finds most important
# We'll look at the magnitude of weights in the first layer
first_layer_weights = model.layers[0].get_weights()[0]  # Shape: (7, 4)
feature_importance = np.mean(np.abs(first_layer_weights), axis=1)

print("📊 Feature Importance (based on first layer weights):")
for i, (name, importance) in enumerate(zip(feature_names, feature_importance)):
    print(f"   {i+1}. {name:<8}: {importance:.3f}")

# Plot feature importance
plt.figure(figsize=(10, 6))
bars = plt.bar(feature_names, feature_importance, color='skyblue', alpha=0.7)
plt.title('Feature Importance in Neural Network')
plt.xlabel('Features')
plt.ylabel('Average Absolute Weight')
plt.xticks(rotation=45)

# Add value labels on bars
for bar, importance in zip(bars, feature_importance):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
             f'{importance:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# ============================================================================
# STEP 8: COMPARISON WITH PLAYGROUND RESULTS
# ============================================================================

print("\n=== COMPARISON WITH TENSORFLOW PLAYGROUND ===")
print("TensorFlow Playground Results:")
print("   ✓ Training Loss: 0.010")
print("   ✓ Test Loss: 0.023") 
print("   ✓ Architecture: [4, 5, 5] neurons")
print("   ✓ Learning Rate: 0.03")
print("   ✓ Activation: Sigmoid")
print("   ✓ Features: All 7 features")

print(f"\n Python Implementation Results:")
print(f"   📊 Training Loss: {train_loss:.3f}")
print(f"   📊 Test Loss: {test_loss:.3f}")
print(f"   📊 Training Accuracy: {train_accuracy:.3f}")
print(f"   📊 Test Accuracy: {test_accuracy:.3f}")

# Performance comparison
if test_loss <= 0.05:
    print("🎯 EXCELLENT: Matched or exceeded playground performance!")
elif test_loss <= 0.1:
    print("✅ GOOD: Close to playground performance")
else:
    print("📈 LEARNING OPPORTUNITY: Try adjusting hyperparameters")

print(f"\n🧠 KEY INSIGHTS:")
print(f"   • Polynomial features (X₁², X₂², X₁X₂) are crucial for spiral patterns")
print(f"   • Feature scaling is essential for neural network training")
print(f"   • The architecture balance prevents overfitting while capturing complexity")
print(f"   • Playground intuition translated well to real TensorFlow code!")

# ============================================================================
# STEP 9: HYPERPARAMETER EXPERIMENTATION
# ============================================================================

print("\n--- Hyperparameter Experimentation Tips ---")
print("Based on playground success, try these variations:")
print("1. Different activations: ReLU, tanh (change 'sigmoid' above)")
print("2. Learning rates: 0.01, 0.1, 0.3 (modify learning_rate=0.03)")
print("3. Architecture: [6,4,4], [8,6,4] (modify Dense layer sizes)")
print("4. Regularization: Add Dropout layers for overfitting")
print("5. More epochs: Increase epochs if loss is still decreasing")

print("\n🎉 CONGRATULATIONS!")
print("You've successfully translated your TensorFlow Playground intuition")
print("into real Python/TensorFlow code. This is exactly how ML works in practice!")
print("You understand: feature engineering, architecture design, hyperparameter tuning!")

# Optional: Save the model
model.save('spiral_classifier.keras')
print(f"\n💾 Model saved as 'spiral_classifier.keras'")
print("It can be downloaded with: tf.keras.models.load_model('spiral_classifier.keras')")
```
